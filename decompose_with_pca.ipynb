```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('temp.npy')
data.shape
```

```python
data = data[:,0,-1,:,:]
```

```python
data.shape
```

```python
data = data.reshape(501,-1)
data.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(data)
```

```python
plt.plot(pca.explained_variance_ratio_,'-o')
plt.xlim(0,20)
```

```python
print(pca.explained_variance_ratio_[:10].sum())
```

```python
decomposed_data = pca.transform(data)[:,:10]
```
